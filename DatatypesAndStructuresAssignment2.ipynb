4.Describe how dictionaries store data?
->A dictionary in Python is an unordered, mutable, and indexed collection of key-value pairs.
Python dictionaries are implemented using a hash table — a data structure that provides fast access to values based on their keys.
1. Hashing the Key
When you add a key-value pair, Python hashes the key using Python’s built-in hash() function.

The result is a hash value (a number), used to determine where in memory the value should go.


```python
hash("name")
```

2. Finding an Index
The hash value is used to find an index in an internal array (like a bucket).

Python uses this index to store the key-value pair.
3. Handling Collisions
Sometimes, different keys can produce the same hash (a collision).

Python handles this using a technique called open addressing (e.g., probing for the next empty slot).

4. Retrieving a Value
To retrieve a value like my_dict["name"]:

Python hashes the key.

It uses the hash to find the index.

It checks if the key at that index matches.

If yes, it returns the associated value.

8.What is a hash table, and how does it relate to dictionaries in Python?
->A hash table is a data structure that maps keys to values using a process called hashing.
Hashing: A function turns a key (like "name" or 123) into a number called a hash.

Indexing: That number is used as an index to store the value in a table (usually an array).

Fast lookup: This makes finding a value by its key very fast — typically O(1) time.
How Does This Relate to Dictionaries in Python?
In Python, the dictionary (dict) is implemented using a hash table behind the scenes.

my_dict = {"name": "Nikita"}
Hashing: Python runs hash("name") to get a unique number.

Indexing: That number is used to find a spot in memory to store "Nikita".

Storing: The key and value are stored together in that spot.

Lookup: When you do my_dict["name"], Python hashes "name" again, finds the index, and returns "Nikita".

17.Describe the time complexity of accessing elements in a dictionary?
->In Python, dictionaries are implemented using hash tables, which provide fast access to elements. The time complexity of accessing elements in a dictionary is generally O(1), meaning it takes constant time to retrieve a value associated with a given key, regardless of the size of the dictionary.

Here's why:
Hashing: When you use a key to access an element in a dictionary, Python computes the hash of the key (using a hash function) and uses that hash to directly locate the corresponding value in the dictionary. This lookup operation is typically done in constant time.

Average Case (O(1)): Under normal circumstances, accessing a value using a key (i.e., dict[key]) has a time complexity of O(1). This is because the hash table is designed to allow quick access, and the hash function typically distributes keys evenly across the available slots.

Worst Case (O(n)): In rare cases, when there are hash collisions (i.e., multiple keys that produce the same hash value), Python handles collisions by using techniques like chaining (linking elements in the same slot). In such cases, the worst-case time complexity can degrade to O(n), where n is the number of elements in the dictionary. However, Python’s implementation of dictionaries is highly optimized to minimize collisions, so this worst-case scenario is rare.

18.In what situations are lists preferred over dictionaries?
->While dictionaries offer fast lookups and are ideal for situations where you need to store key-value pairs, lists have their own advantages in specific scenarios. Lists are generally preferred over dictionaries when:

1. You Need an Ordered Collection of Items:
When order matters: Lists maintain the order of insertion, meaning elements are stored in the sequence in which they are added. If you need to preserve the order of items or iterate through them in a specific sequence, a list is ideal.
Example: Storing a sequence of user inputs, tasks to be completed in order, or ordered data (like a timeline of events).

2. When You Need to Store an Indexed Collection of Items:
Accessing by index: Lists are better when you need to access elements by their index or position in the sequence. You can use integer indices to refer to each item in the list, making it easier to iterate over or modify specific elements.

3.ou Need to Store Simple, Homogeneous Data:
When the data type is uniform: Lists are best when you're working with a collection of similar items (e.g., integers, strings, or objects). Since dictionaries store key-value pairs, they are more complex and introduce unnecessary overhead when you don’t need key-value relationships.

4. When You Need to Perform Operations Like Sorting or Filtering:
Easy to manipulate data: Lists have built-in methods like .sort(), .reverse(), and list comprehensions, which allow you to easily filter, sort, or modify the data. These operations are often more natural and efficient with lists compared to dictionaries.



19.Why are dictionaries considered unordered, and how does that affect data retrieval?
->In Python, dictionaries were traditionally considered unordered because they did not maintain any specific order for the key-value pairs. However, this changed starting with Python 3.7, where dictionaries guarantee the insertion order. Despite this, dictionaries are still technically considered unordered collections in the sense that their design is optimized for fast lookups and not for preserving a sequence of elements based on their order.
Hashing:

Dictionaries in Python are implemented using hash tables. When you insert a key-value pair into a dictionary, the key is hashed, and the corresponding value is placed in the "bucket" based on that hash.

The position of an item in the dictionary is determined by its hash value, not its order of insertion. This allows for fast lookups but means the internal structure is not organized in any sequential order.

Insertion Order Guarantee:

dictionaries do maintain insertion order. This means that when you iterate over the dictionary, the items will be returned in the order they were added. However, this does not mean that dictionaries are "ordered" in the same way that lists are, since the primary purpose of a dictionary is efficient data retrieval, not order preservation.

This insertion order guarantee is more of a side effect of the internal structure, rather than a core characteristic of dictionaries.
